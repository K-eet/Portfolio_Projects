```python
# 02_visualization.ipynb
# Visualizing financial metrics: Tesla vs BYD vs Ford

import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pickle

# Load metrics from data collection notebook
with open('../data/metrics.pkl', 'rb') as f:
    metrics = pickle.load(f)

print("Metrics loaded:")
for name in metrics:
    print(f"{name}:")
    print(metrics[name])
    print()
```

```python
# Define consistent colors for each company
colors = {
    "Tesla": "#E31937",   # Tesla red
    "BYD": "#1DB954",     # BYD green
    "Ford": "#003DA5"     # Ford blue
}

# Net Profit Margin comparison
fig = go.Figure()

for name in metrics:
    fig.add_trace(go.Scatter(
        x=metrics[name].index.year,
        y=metrics[name]["Net Profit Margin (%)"],
        name=name,
        mode="lines+markers",
        line=dict(color=colors[name], width=2),
        marker=dict(size=8)
    ))

fig.update_layout(
    title="Net Profit Margin (%) — Tesla vs BYD vs Ford (2022–2024)",
    xaxis_title="Year",
    yaxis_title="Net Profit Margin (%)",
    legend_title="Company",
    hovermode="x unified",
    template="plotly_white"
)

fig.update_xaxes(
    tickmode="array",
    tickvals=[2022, 2023, 2024],
    ticktext=["2022", "2023", "2024"]
)

fig.show()
```

```python
# Free Cash Flow comparison
fig2 = go.Figure()

for name in metrics:
    fig2.add_trace(go.Scatter(
        x=metrics[name].index.year,
        y=metrics[name]["Free Cash Flow"] / 1e9,  # Convert to billions
        name=name,
        mode="lines+markers",
        line=dict(color=colors[name], width=2),
        marker=dict(size=8)
    ))

fig2.update_layout(
    title="Free Cash Flow (USD Billions) — Tesla vs BYD vs Ford (2022–2024)",
    xaxis_title="Year",
    yaxis_title="Free Cash Flow (USD Billions)",
    legend_title="Company",
    hovermode="x unified",
    template="plotly_white"
)

fig2.update_xaxes(
    tickmode="array",
    tickvals=[2022, 2023, 2024],
    ticktext=["2022", "2023", "2024"]
)

fig2.show()

# Return on Assets comparison
fig3 = go.Figure()

for name in metrics:
    fig3.add_trace(go.Scatter(
        x=metrics[name].index.year,
        y=metrics[name]["Return on Assets (%)"],
        name=name,
        mode="lines+markers",
        line=dict(color=colors[name], width=2),
        marker=dict(size=8)
    ))

fig3.update_layout(
    title="Return on Assets (%) — Tesla vs BYD vs Ford (2022–2024)",
    xaxis_title="Year",
    yaxis_title="Return on Assets (%)",
    legend_title="Company",
    hovermode="x unified",
    template="plotly_white"
)

fig3.update_xaxes(
    tickmode="array",
    tickvals=[2022, 2023, 2024],
    ticktext=["2022", "2023", "2024"]
)

fig3.show()
```
