```python
import requests
import json

# SEC requires a descriptive User-Agent header with contact info
HEADERS = {
    "User-Agent": "Personal Portfolio Project keet@example.com"
}

# CIK numbers (need to verify these)
cik_numbers = {
    "Tesla": "1318605",
    "Ford": "37996"
}

def get_company_facts(cik):
    """Fetch all XBRL facts for a company by CIK."""
    cik_padded = str(cik).zfill(10)
    url = f"https://data.sec.gov/api/xbrl/companyfacts/CIK{cik_padded}.json"
    response = requests.get(url, headers=HEADERS)
    response.raise_for_status()
    return response.json()

# Test pull for Tesla
tesla_facts = get_company_facts(cik_numbers["Tesla"])
print(f"Company name: {tesla_facts['entityName']}")
print(f"Available concepts: {len(tesla_facts['facts']['us-gaap'])}")
```
